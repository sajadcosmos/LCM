```
import numpy as np
import matplotlib.pyplot as plt
#from matplotlib import pyplot as plt
%matplotlib inline
```

```
y = np.array([])
limit =301
for k in range (1, limit):
  c = 0
  for i in range (1, k+1):
    for j in range (1, k+1):
      if np.lcm(i,j) == k:
        c = c+1
  y = np.append(y, c/(k**2))

x = np.arange(1, limit)
plt.plot(x, y)
plt.title("Variation of fractional number  of ordered pairs with n")
plt.xlabel("Value of n")
plt.ylabel("f(n)/nÂ²")
plt.show()
```

```
w=[]
for i in range(1, limit):
  w.append(np.average(y[:i]))
```

```
plt.plot(x, w)
plt.title("Variation of moving average")
plt.xlabel("Value of n")
plt.ylabel("moving average")
plt.show()
```

```
vary=[]
for i in range(1,limit):
  vary.append(np.var(y[:i]))
```

```
plt.plot(x, vary)
plt.title("Variation of moving variance")
plt.xlabel("Value of n")
plt.ylabel("moving average variance ")
plt.show()
```
